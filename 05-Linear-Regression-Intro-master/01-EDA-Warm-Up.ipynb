```python
import numpy as np
pd.pivot_table(df2, values = 'Number of students who registered for the SHSAT', index = 'School name', columns = 'Year of SHST', aggfunc=np.mean)
```

```python
reg_table.loc[reg_table.index[10]].sort_values(ascending = False).plot(kind = 'bar', title = reg_table.index[10], color = ['lightblue', 'grey', 'grey', 'grey'] )
```

```python
def part_plot(school_name):
    print(reg_table.loc[school_name])
    return reg_table.loc[school_name].sort_values(ascending = False).plot(kind = 'bar', title = school_name, color = ['lightblue', 'grey', 'grey', 'grey'], figsize = (10, 4) )
```

```python
reg_table.index
```
