```python
import pandas as pd
import geopandas
```

```python
df = pd.read_csv('gun-violence-data_01-2013_03-2018.csv (2).zip', compression='zip')
```

```python
df.info()
```

```python
import folium
```

```python
m = folium.Map(location=[df.latitude.mean(), df.longitude.mean()], zoom_start=4, tiles="CartoDB dark_matter")
```

```python
m
```

```python
for index, row in df.dropna().iterrows():
    folium.CircleMarker(location=(row["latitude"],
                                  row["longitude"]),
                        radius=row["n_killed"],
               fill=True).add_to(m)
```

```python
m
```

```python
m.save('test.png')
```
